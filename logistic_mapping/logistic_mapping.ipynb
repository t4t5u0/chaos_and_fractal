```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 初期値
a = 3.2
x_0 = 0.1
```

```python
# x_n+1 = ax_n(1-x_n)
def logistic(a, x_0):
    x = [x_0]
    for _ in range(100):
        x.append(a * x[-1] * (1-x[-1]))
    return x
```

```python
# 演算
x = logistic(a, x_0)

# リターンマップ
fig = plt.figure() 
ax = fig.add_subplot(xlabel='x_n', ylabel='x_n+1', title='return map')
ax.plot(x[:len(x)-1], x[1:],ls='', marker='.')
plt.show()
```

```python
x = logistic(a, x_0)

# 時系列グラフ
fig = plt.figure() 
ax = fig.add_subplot(xlabel='n', ylabel='x_n', title='time series')
ax.plot(x)
plt.show()
```
